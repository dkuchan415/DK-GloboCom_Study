```python
#Load csv file

file = "Globo-ComCaseStudyData.csv"
globo_df = pd.read_csv(file)
globo_df.head()
```

```python
#Create TotalCharge field

globo_df["TotalCharge"] = globo_df["Day Charge"] + globo_df["Eve Charge"] + globo_df["Night Charge"] + globo_df["Intl Charge"]

globo_df.head()

#Create TotalCalls field

globo_df["TotalCalls"] = globo_df["Day Calls"] + globo_df["Eve Calls"] + globo_df["Night Calls"] + globo_df["Intl Calls"]

#Create TotalMins field

globo_df["TotalMins"] = globo_df["Day Mins"] + globo_df["Eve Mins"] + globo_df["Night Mins"] + globo_df["Intl Mins"]

globo_df.head()
```

```python
#Change churn 1 and 0 to Yes and No

globo_df["Churn?"].replace({1:"Yes", 0:"No"}, inplace=True)
globo_df.head()
```

```python
#Determine overall churn %

globo_df.loc[globo_df["Churn?"] == "Yes"].count()/globo_df["Churn?"].count()
```

```python
globo_df["TotalCharge"].describe()
```

```python
#Create histogram for all customers
sns.set_style("darkgrid")
sns.color_palette("dark")
both_hist = sns.histplot(data=globo_df, x="TotalCharge", hue="Churn?", color="g", multiple="stack", bins=10, palette="dark")
plt.ylabel("Customer Count")
plt.xlabel("Customer Spend")
plt.title("Customer Churn by Spend Range")
plt.savefig("churn_hist.png")
plt.show()
```

```python
#Create df for churn customers

churn_df = globo_df.loc[globo_df["Churn?"] == "Yes"]
churn_df.head()
```

```python
#Create df for non churn customers

non_churn_df = globo_df.loc[globo_df["Churn?"] == "No"]
non_churn_df.head()
```

```python
#Create bins for total spend

bins = [20, 27.5, 35, 42.5, 50, 57.5, 65, 72.5, 80, 87.5, 95]
group_names = ["20-27.50", "27.50-35", "35-42.50", "42.50-50", "50-57.50", "57.50-65", "65-72.50", "72.50-80", "80-87.50", "87.50-95"]

globo_df["SpendRange"] = pd.cut(globo_df["TotalCharge"], bins, labels=group_names, include_lowest=True)
globo_df

#Create overall spend groupings

spend_group = globo_df.groupby("SpendRange")

spend_group_count = spend_group.count()
spend_group_count

spend_group_sum = spend_group.sum()

spend_total = spend_group_sum["TotalCharge"].sum()

spend_percent = spend_group_count["TotalCharge"]/spend_group_count["TotalCharge"].sum()

spend_percent_df = pd.DataFrame({"Customer Count": spend_group_count["TotalCharge"],
                               "Spend Percent": spend_percent,
                               "Total Spend": spend_group_sum["TotalCharge"]})
spend_percent_df
```

```python
# Create bins for Churn df spend range

churn_df["SpendRange"] = pd.cut(churn_df["TotalCharge"], bins, labels=group_names, include_lowest=True)
churn_df

churn_group = churn_df.groupby("SpendRange")

churn_group_count = churn_group.count()
churn_group_count

churn_count_df = pd.DataFrame({"Churn Count": churn_group_count["TotalCharge"]})
churn_count_df
```

```python
churn_group_spend = churn_group.sum()
churn_group_spend

churn_spend_df = pd.DataFrame({"Churn Spend": churn_group_spend["TotalCharge"]})
churn_spend_df
```

```python
churn_pct_df = spend_percent_df.merge(churn_count_df,how="left",on="SpendRange")
churn_pct_df

```

```python
#Create 
#.apply('{:.0%}'.format)

churn_pct_df["Churn Percent"] = (churn_pct_df["Churn Count"]/churn_pct_df["Customer Count"])
# churn_pct_df
churn_pct_df
```

# Marketing


```python
#Create Int'l df

intl_df = globo_df.loc[globo_df["Int'l Plan"] == "yes"]
intl_df["IntlSpend%"] = (intl_df["Intl Charge"]/intl_df["TotalCharge"])*100
intl_df.head()
```

```python
intl_df["TotalCharge"].describe()
```

```python
#Create boxplots

sns.set_style("darkgrid")
sns.catplot(y="IntlSpend%", data=intl_df, kind="box", x="Churn?")
plt.ylabel("International Spend %")
plt.title("Int'l Spend in Churn vs. Non-Churn Customers")
plt.gcf().subplots_adjust(bottom=0.15)
plt.savefig("intl_churn.png")
plt.show()

#Count Int'l customers from total

intl_group = globo_df.groupby("Int'l Plan")
intl_group.count()

# sns.catplot(x="Intl Calls", y="TotalCalls", data=intl_df, hue="Churn?")
# plt.show()
```

```python
# apply to globo df
globo_df["IntlSpend%"] = (globo_df["Intl Charge"]/globo_df["TotalCharge"])*100
globo_df.head()
```

```python
# Calculate total churn %

intl_churn_df = intl_df.loc[globo_df["Churn?"] == "Yes"]
intl_churn_df["TotalCharge"].sum()/churn_df["TotalCharge"].sum()
```
